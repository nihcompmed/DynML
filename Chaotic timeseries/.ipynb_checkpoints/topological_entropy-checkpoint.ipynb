```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.integrate import solve_ivp
import seaborn as sns
import time

# ================== CONFIGURATIONS ======================
np.random.seed(42)
param_range_reservoir = [(9, 11), (2.6, 2.8), (27, 29)]  # (sigma, rho, beta) ranges
param_range_R = [(1, 30)]  # Range for R matrix random values
t_initial, t_final = 0, 1  # Integration time span
num_eval_points = 500  # Reduce time points for speed
input_csv = "input_data.csv"
output_csv = "output_data.csv"

# Epsilon values for entropy calculations
eps_values = [0.9, 0.5, 0.1, 0.05, 1e-2, 1e-4, 1e-6, 1e-10]

# ============ LOAD AND NORMALIZE TRAINING DATA ===========
num_train_samples = 1000
input_data = pd.read_csv(input_csv).iloc[:, 1:num_train_samples+1].values
output_data = pd.read_csv(output_csv).iloc[:, 1:num_train_samples+1].values

D, M, P = input_data.shape[1], input_data.shape[0], output_data.shape[0]

scaler_input = StandardScaler()
input_data_norm = scaler_input.fit_transform(input_data.T).T

scaler_output = StandardScaler()
output_data_norm = scaler_output.fit_transform(output_data.T).T

# ============ DEFINE LORENZ SYSTEM ==============
def lorenz_system(t, X, params):
    X = X.reshape(len(params), 3)
    dxdt = (1 / params[:, 3]) * params[:, 0] * (X[:, 1] - X[:, 0])  
    dydt = (1 / params[:, 3]) * (X[:, 0] * (params[:, 1] - X[:, 2]) - X[:, 1])
    dzdt = (1 / params[:, 3]) * (X[:, 0] * X[:, 1] - params[:, 2] * X[:, 2])
    dXdt = np.stack([dxdt, dydt, dzdt], axis=1).flatten()
    return dXdt

# ============ COMPUTE ENTROPY ==============
def compute_topological_entropy(trajectory, eps_values):
    entropy_estimates = {}
    for eps in eps_values:
        unique_orbits = set()
        for point in trajectory:
            rounded_point = tuple(np.round(point / eps))  
            unique_orbits.add(rounded_point)
        N_n_eps = len(unique_orbits)
        entropy_estimates[eps] = np.log(N_n_eps) / len(trajectory)  
    return entropy_estimates

# ============ LOOP OVER N =============
N_values = range(10, 310, 10)  # N = 10 to 300 in steps of 10
entropy_results = {eps: [] for eps in eps_values}

start_time = time.time()

for N in N_values:
    print(f"Running for N = {N}...")
    
    # Generate random parameters for reservoirs
    reservoir_params = np.array([
        [np.random.uniform(low, high) for low, high in param_range_reservoir] + [np.random.uniform(1, 10)]
        for _ in range(N)
    ])

    # Build random weight matrix R
    R = np.random.uniform(param_range_R[0][0], param_range_R[0][1], size=(3 * N, M))

    # Initialize entropy sum for each epsilon
    entropy_sums = {eps: 0 for eps in eps_values}
    
    for i in range(D):
        # Input mapping
        input_vector = np.dot(R, input_data_norm[:, i]).flatten()

        # ODE Solve
        t_eval = np.linspace(t_initial, t_final, num_eval_points)
        sol = solve_ivp(
            lorenz_system, (t_initial, t_final), input_vector, 
            t_eval=t_eval, args=(reservoir_params,), method='RK45', rtol=1e-6, atol=1e-9
        )
        
        # Compute entropy
        trajectory = sol.y.T
        entropy_values = compute_topological_entropy(trajectory, eps_values)
        
        # Sum entropy values for averaging later
        for eps in eps_values:
            entropy_sums[eps] += entropy_values[eps]

    # Compute average entropy over all D iterations
    for eps in eps_values:
        entropy_results[eps].append(entropy_sums[eps] / D)


# ============ SAVE RESULTS TO CSV =============

# Create a DataFrame with N as the first column
entropy_df = pd.DataFrame({
    "N": list(N_values)
})

# Add entropy columns for each epsilon
for eps in eps_values:
    entropy_df[f"entropy_eps_{eps}"] = entropy_results[eps]

# Save to CSV
output_csv_file = "entropy_vs_N.csv"
entropy_df.to_csv(output_csv_file, index=False)

print(f"Entropy results saved to {output_csv_file}")

end_time = time.time()
print(f"Total Execution Time: {end_time - start_time:.2f} seconds")
```

```python
# ================= LOAD RESULTS =================
entropy_df = pd.read_csv("entropy_vs_N.csv")

# Extract N values
N_values = entropy_df["N"].values

# Identify epsilon columns automatically
eps_columns = [col for col in entropy_df.columns if col.startswith("entropy_eps_")]

# ================== PLOT ==================
plt.figure(figsize=(6, 4))

for col in eps_columns:
    # Extract epsilon value from column name
    eps = col.replace("entropy_eps_", "")
    
    plt.plot(
        N_values,
        entropy_df[col].values,
        linewidth=2,
        label=rf"$\epsilon = {eps}$"
    )

plt.xlabel("Number of Reservoirs ($N$)", fontsize=12)
plt.ylabel("Average Topological Entropy", fontsize=12)
plt.title("Scaling of Reservoir Entropy with System Size", fontsize=14)

plt.legend(fontsize=11, frameon=False)
plt.grid(True, which="both", linestyle="--", linewidth=0.6, alpha=0.7)

plt.tight_layout()

# ================= SAVE FIGURE =================
plt.savefig(
    "avg_topological_entropy_vs_N.jpeg",
    format='jpeg',
    dpi=650,
    bbox_inches="tight"
)

plt.show()
```
